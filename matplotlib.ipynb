```
import matplotlib.pyplot as plt # para poder ver graficas
import numpy as np  # funciones numéricas (arrays, matrices, etc.)

fig, ax = plt.subplots()  # Crear una figura que contenga un solo eje..
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Graficar algunos datos en los ejes (x,y)

# tambien se puede crear sin necesidad de crear el eje fig con plt.plot([x], [y])
```

```
# de la forma orientada a objetos (O,O)
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  #  Crear una figura y un eje.
ax.plot(x, x, label='linear')  # Trace algunos datos en los ejes.
ax.plot(x, x**2, label='quadratic')  # Trazar más datos sobre los ejes ...
ax.plot(x, x**3, label='cubic')  # ...  y algo más.
ax.set_xlabel('x label')  # Agregue una etiqueta x a los ejes..
ax.set_ylabel('y label')  # Agregue una etiqueta y a los ejes
ax.set_title("Simple Plot")  # Agregar un título a los ejes.
ax.legend()  # Agregar una leyenda.
```

```
# De la forma (pyplot-style)
# se pueden combinar las dos formas. para titulos, leyendas y demas.
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Trace algunos datos en los ejes (implícitos).
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')  # Agregue una etiqueta x a los ejes
plt.ylabel('y label')  # Agregue una etiqueta y a los ejes
plt.title("Simple Plot")
plt.legend()
```

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# Setup, and create the data to plot
y = np.random.rand(100000)
y[50000:] *= 2
y[np.logspace(1, np.log10(50000), 400).astype(int)] = -1
# path.simplify es parámetro es un valor booleano que indica si los segmentos de línea están simplificados o no
mpl.rcParams['path.simplify'] = True

mpl.rcParams['path.simplify_threshold'] = 0.0
plt.plot(y)
plt.show()

# path.simplify_threshold es un parámetro controla la cantidad de segmentos de línea simplificados
mpl.rcParams['path.simplify_threshold'] = 1.0 
plt.plot(y)
plt.show()
```

```
# funcion seno
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)   # valores inicio , fin , saltos
s = 1+ 2*np.sin(2 * np.pi * t)  # funcion f(t)

fig, ax = plt.subplots() #  Crear una figura y un eje
ax.plot(t, s)  # traza algunos datos en los ejes.

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title="Grafico de $1+sin(2* \pi*t)$ ")
ax.grid()

fig.savefig("test.png") # guarda la figura
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(0.0, 6.0) # grafica 1 eje x
x2 = np.linspace(0.0, 2.0) # grafica 2 eje x

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1) # funcion 1
y2 = np.cos(2 * np.pi * x2) # funcion 2

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-') # como se grafica que contra que
plt.title('A tale of 2 subplots')
plt.ylabel('oscilacion amortiguada') # para grafica 1

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')  # como se grafica par la grafica 2
plt.xlabel('time (s)') 
plt.ylabel('Undamped') # para grafica 2

plt.show()
```

```
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Hace los datos.
X = np.arange(-6, 6, 0.25) # valores inicio , fin , saltos
Y = np.arange(-5, 5, 0.25) # valores inicio , fin , saltos
X, Y = np.meshgrid(X, Y) 
R = np.sqrt(X**2 + Y**2)  # funcion f(x,y)
Z = np.sin(R)  # funcion f(z)

# Traza la superficie.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Personaliza el eje z.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10)) # llas diviciones del eje z
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Agregar una barra de colores que asigna valores a los colores.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
