```
import pygame
 
def main():
 
    pygame.init()
 
    screen = pygame.display.set_mode( (500,500) )
 
    background = pygame.Surface( screen.get_size() )
 
    background.fill( (0,0,0) )
 
    image = pygame.image.load("Alexandra17.jpg")
 
    image_position = image.get_rect()
 
    image_position.bottom = 500
    image_position.left = 0
 
    screen.blit( background, (0,0) )
    screen.blit( image, image_position )
 
    pygame.display.flip()
 
    while 1:
        pygame.event.pump()
        keyinput = pygame.key.get_pressed()
 
        if keyinput[pygame.K_ESCAPE] or pygame.event.peek( pygame.QUIT ):
            break
 
        screen.blit( background, (0,0) )
        screen.blit( image, image_position )
 
        pygame.display.flip()
 
if __name__ == '__main__':
  
  main()
```

```
from PIL import Image, ImageChops, ImageEnhance, ImageOps
def main():
    image = Image.open("Alexandra17.jpg")
    
    # Invertir colores.
    new_image = ImageChops.invert(image)
    new_image.save("Alexandra17.jpg")
    
    # Escala de grises.
    new_image = ImageOps.grayscale(image)
    new_image.save("Alexandra17.jpg")
    
    # Resaltar luces.
    new_image = ImageEnhance.Brightness(image).enhance(2)
    new_image.save("Alexandra17.jpg")
    
    # Contraste.
    new_image = ImageEnhance.Contrast(image).enhance(4)
    new_image.save("Alexandra17.jpg")
    
    # Espejo.
    new_image = ImageOps.mirror(image)
    new_image.save("Alexandra17.jpg")
    
    # Cambiar tamaño.
    new_image = image.resize((320, 240))
    new_image.save("Alexandra17.jpg")
    
    # Diminuir nitidez.
    new_image = ImageEnhance.Sharpness(image).enhance(-4)
    new_image.save("Alexandra17.jpg")
if __name__ == "__main__":
main()
```
