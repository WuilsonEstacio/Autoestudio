```
# Para tabular una funcion en terminos de x 
def add_five(x):
  return x+5
nums_x=[11,22,33,44,55,67]
result=list(map(add_five,nums_x))
print(result)
```

```
# Para tabular una funcion en terminos de x y 
def add_five(x,y):
  return x*y+5
nums_x=[11,22,33,44,55,67]
nis_y=[2,3,4,5,6,8]
result=list(map(add_five,nums_x,nis_y))
print(result)
```

```
def power(x,y):
  if y==0:
    return 1
  else:
    return x*power(x,y-1)
print(power(2,3))
```
