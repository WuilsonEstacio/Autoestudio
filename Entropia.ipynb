```
import numpy as np
import matplotlib.pyplot as plt
```

## Función para mover las partículas

```
np.random.seed(42)
def mover(xx):
  rand = np.random.uniform(0, 1.25)
  
  if (rand >= 0) and  (rand < 0.25) :
    xx[0] =  xx[0] + 1
  if (rand >= 0.25) and  (rand < 0.50) :
    xx[0] =  xx[0] - 1
  if (rand >= 0.50) and  (rand < 0.75) :
    xx[1] =  xx[1] + 1
  if (rand >= 0.75) and  (rand <= 1.0):
    xx[1] =  xx[1] - 1
  else:
    xx = xx  
  return xx
```

# Condiciones de Frontera

```
def frontera(xx):
  if xx[0] > 100 :
      xx[0] =  xx[0] - 1
  if xx[1] > 100 :
      xx[0] =  xx[1] - 1
  if xx[0] < -100 :
      xx[0] =  xx[0] + 1
  if xx[1] < -100:
      xx[1] =  xx[1] + 1
  return xx
```

```
# Número de partículas
N = 400

# Iniciamos las coordenadas de las partíclas
particulas = []
for i in range(int(np.sqrt(N))):
  for j in range(int(np.sqrt(N))):
    particulas.append([i -10 , j -10 ])
# se mueven las partículas en tiempo t
t = 10**4 + 1
# tiempo para imprimir la posición de las partículas
t_print = [0, 10**1, 10**2, 10**3, 10**4]




for i in range(t):
  for j in range(len(particulas)):
    if i in t_print:
      print('Tiempo:', i)
      # Se alistan los datos para graficarlos
      x = []
      y = []
      for i in range(len(particulas)):
        x.append(particulas[i][0])
        y.append(particulas[i][1])

      plt.figure(figsize=(8, 8))
      plt.scatter(x, y, color='black')
      plt.axis(xmin = -100, xmax = 100, ymin = -100, ymax = 100)  
      plt.grid()
      plt.xlabel('x')
      plt.ylabel('y')
      plt.show()



    particulas[j] =  mover(particulas[j])
    particulas[j] =  frontera(particulas[j])

```
