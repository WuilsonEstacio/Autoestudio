```
# la sentencia else utilizada dentro del bucle for o while se ejecuta si el bucle es normal
for i in range(2,10):
  if i>5:
    print(i)
    break
else:
  print("7") 
```

```
# la sentencia else utilizada en try  se ejecuta solo si no hay error alguno
try:
  print(1)
except ZeroDivisionError:
  print(2)
else:
  print(3)

try:
  print(1/0)
except ZeroDivisionError:
    print(4)
else:
  print(5)
```

```
try:
  print(1)
  print(1+"1"==2)
  print(2)
except TypeError:
  print(3)
else:
  print(4)
```

```
# Beautiful Soup analiza el documento HTML dado en un árbol de objetos Python.
# Hay cuatro objetos Python principales de los que necesitas saber: Tag, 
# NavigableString, BeautifulSoup, y Comment.
# https://code.tutsplus.com/es/tutorials/scraping-webpages-in-python-with-beautiful-soup-the-basics--cms-28211
from bs4 import BeautifulSoup

soup = BeautifulSoup("<html><p>This is <b>invalid HTML</p></html>", "html.parser")
print(soup)

soup = BeautifulSoup("<html><p>This is <b>invalid HTML</p></html>", "lxml")
print(soup)
 
soup = BeautifulSoup("<html><p>This is <b>invalid HTML</p></html>", "xml")
print(soup)
 
soup = BeautifulSoup("<html><p>This is <b>invalid HTML</p></html>", "html5lib")
print(soup)
```

```
import requests
from bs4 import BeautifulSoup
 
req = requests.get('https://en.wikipedia.org/wiki/Python_(programming_language)')
soup = BeautifulSoup(req.text, "lxml")

soup.title  # para el titulo

soup.title.name  # para el nombre

soup.title.string  # 

soup.h1  # 
 
soup.h1.string   # 

soup.h1['class']  # para acceder a las clases

soup.h1['id']   #  para id y etiquetas

soup.h1.attrs   # para acceder al diccionario completo
# 
soup.h1['class'] = 'firstHeading, mainHeading'
soup.h1.string.replace_with("Python - Programming Language")
del soup.h1['lang']
del soup.h1['id']
 
soup.h1

# De manera similar, puedes iterar a través de todos los enlaces o 
# sub-encabezados en un documento usando el siguiente código

for sub_heading in soup.find_all('h2'):
    print(sub_heading.text)


print(soup.p.contents) # acceder a todos los hijos de una etiqueta

 # puedes obtener el primer enlace en el primer párrafo de la página  usando soup.p.
for child in soup.p.children: # puedes iterar sobre el hijo de una etiqueta
    print(child.name)

print(soup.p.parent.name)
 
for parent in soup.p.parents:
    print(parent.name)

```
