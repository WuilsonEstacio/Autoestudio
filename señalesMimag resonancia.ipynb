```
import numpy as np
import random
import matplotlib.pylab as p
from scipy.fftpack import fft, fftfreq

#Paso 1: Generar senal y hacer TF

Mo=1
w=42.6*2*np.pi #MHz
B=1 #T
T2=100 #ms
t=np.linspace(0,100,1000)
x=Mo*np.sin(w*t)*np.exp(-t/T2)
y=Mo*np.cos(w*t)*np.exp(-t/T2)
M=x+y

###Transformada de Fourier

F=fft(M)
Ft=np.abs(F)
g=t/(2*np.pi)


fig2, ax2 = p.subplots()
ax2.plot(g,Ft)
fig, ax = p.subplots()
ax.plot(t,M)
p.show()
#Paso 2: Generar senal con variacion de frecuencias, a1 y a2 igual a 1
a1=1
a2=1

for i in range(10,100,6):
	for j in range(20,110,5):
		L=Mo*(a1*np.sin(i*t)+a2*np.sin(j*t))*np.exp(-t/T2)
		R=np.abs(fft(L))
	p.plot(t,L)
	#p.plot(g,R)
	p.show()

for i in range(10,100,6):
	for j in range(20,110,5):
		L=Mo*(a1*np.sin(i*t)+a2*np.sin(j*t))*np.exp(-t/T2)
		R=np.abs(fft(L))
	#p.plot(t,L)
	p.plot(g,R)
	p.show()

#Paso 3: Generar senal con variacion de frecuencias, a1 y a2 diferentes a 1
a1=3
a2=4

for i in range(10,100,6):
	for j in range(20,110,5):
		L=Mo*(a1*np.sin(i*t)+a2*np.sin(j*t))*np.exp(-t/T2)
	p.plot(t,L)
	p.show()

for i in range(10,100,6):
	for j in range(20,110,5):
		L=Mo*(a1*np.sin(i*t)+a2*np.sin(j*t))*np.exp(-t/T2)
		R=np.abs(fft(L))
	p.plot(g,R)
	p.show()
	
#Paso 4: Agregar ruido a la senal

H=0.2*Mo*random.random()
MMx=Mo*np.sin(w*t)*np.exp(-t/T2)+H
FM=np.abs(fft(MMx))


fig5, ax5= p.subplots()
ax5.plot(g,FM)
fig4, ax4= p.subplots()
ax4.plot(t,MMx)
p.show()
```
