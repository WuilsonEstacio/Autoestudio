```
# primero ingresar a https://myaccount.google.com/lesssecureapps y permitir acceso

from smtplib import SMTP
import smtplib # libreria para enviar correos
from decouple import config  # es para que no se vea la contrase√±a y para ahorrar pasos


message = 'Holla, esto es un mensaje de python' # mensaje a trasnmitir
subject = 'prueba de correo'

message = 'Subject: {}\n\n{}'. format(subject,message) # para dar asunto al mensaje
 server = smtplib.SMTP('smtp.gmail.com',587) # aqui se define el servidos y puesto de correo
 server.starttls() # indicamos uso de tls
 server.login('waestacioro@unal.edu.co', config('MAIL_PASSWORD')) # para identificarnos

server.sendmail('waestacioro@unal.edu.co', 'wadolfoe@gmail.com', message) # aqui se coloca el remitente, el destinatario y el mensaje
server.quit() # indicamos al servidor que cerramos secion

print("correo enviado exitosamente") # para confirmar el envio del correo
```
