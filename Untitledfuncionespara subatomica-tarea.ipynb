```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

u = 469.459 # MeVc^2
e = 2.2 # Mev
a = 2.2 # fm
h = 197.3 # MeV

# Función 1.
def f1(x):
  return -np.sqrt(2*u*e)/np.sqrt(2*u*(x-e))

x = np.array(np.array([30.9, 31.1, 31.2, 31.3]))
y = f1(x)
print(y)

tabla1 = pd.DataFrame( list(zip(x, y)), columns=['x', 'f1(x)'] )
```

```
# Función 2.
def f2(x):
    return np.cos(np.sqrt(2*u*(x-e))*(a/h))

x = np.array(np.array([30.9, 31.1, 31.2, 31.3]) )
y = f2(x) 
print(y)

tabla2 = pd.DataFrame( list(zip(x, y)), columns=['x', 'f2(x)'] )
#tabla
```

```
import numpy as np
from matplotlib import pyplot

# Función 1.
def f1(x):
    return -np.sqrt(2*u*e)/np.sqrt(2*u*(x-e))

# Función 2.
def f2(x):
    return np.cos(np.sqrt(2*u*(x-e))*(a/h))

# Valores del eje X que toma el gráfico.
x = range(0, 120)

# Graficar ambas funciones.
pyplot.plot(x, [f1(i) for i in x])
pyplot.plot(x, [f2(i) for i in x])

# Establecer el color de los ejes.
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")

# Limitar los valores de los ejes.
pyplot.xlim(1, 100)
pyplot.ylim(-1, 1)

# valores y titilos de los ejes
plt.ylabel('u(x)')
plt.xlabel('V_0')

# Guardar gráfico como imágen PNG.
pyplot.savefig("output.png")

# Mostrarlo.
pyplot.show()
```
