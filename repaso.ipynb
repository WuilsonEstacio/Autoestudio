```
import threading
import time

n=0

def contador(nombre):
  global n
  print("Inicio: soy el hilo {} y n ={} ".format(nombre, n))
  print("{}: Espera 10 segundos.. .".format(nombre))
  n+= 1
  time.sleep(10)
  print("pasado 10 segundos: soy el hilo = {} y n = {} ".format(nombre, n))

for i in range(10):
  t = threading.Thread(target = contador, args=("H-{}".format(i), ))
  t.start()  
```

```
import threading
import time

n=0

def contador(nombre, i):
  global n
  print("Inicio: soy el hilo {} y n ={} ".format(nombre, n))
  print("{}: Espera 10 segundos...".format(nombre, 10-i))
  time.sleep(10-i)
  n+= 1  
  print("pasado {} segundos: soy el hilo = {} y n = {} ".format(10-i, nombre, n))

for i in range(10):
  t = threading.Thread(target = contador, args=("H-{}".format(i), i ))
  t.start()  
```
