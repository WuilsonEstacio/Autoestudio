```
# Codigo preguntas con while 
respuesta = "si"
i=0
while respuesta =="si":
  i=int(input("Cuanto me quieres de 1 a 50, = "))
  print(" Mi respuesta para ti:")
  if i >= 30:
    print(" He buscado esa señal de tu mirada durante tanto tiempo en mi vida")
    print(" la he buscado en las ovejas, en los libros en los cristales,")
    print(" en el silencio del desierto, y he alli en tus ojos estaba ")
    print(" el puro lenguaje del mundo sin explicaciones porque el universo")
    print(" no necesita explicaciones para continuar su camino,")
    print(" todo lo que entendia es que estaba delante de la mujer mas bella")
    print(" de la mujer de mi vida y sin mas palabras ")
    print(" esperaba que ella lo supiera tambien")
    break
  elif 20 <= i < 30:
    print(" Para todos los caminos \n Que caminaremos juntos, \n Para todos los paisajes")
    print(" Que veremos juntos, \n Un corazon timido y emocionado")
    print(" con todas las preguntas y tambien las respuestas,")
    print(" En todos esos momentos Yo te amo.")
    break
  elif 10 < i < 20:
    print(" Porque ver por separada esta vida y la siguiente")
    print(" si una depende de la anterior")
    break
  elif i==10:
    print("I just want to hold and kiss you from here to eternity.")
    break
  elif 8<=i<10:
    print(" Do you know somethings, your name sounds beautiful when it is accompanied with mine.")
    break
  elif  5<=i < 8:
    print(" Espero que el destino tenga ganas de juntarnos pronto")
    break
  elif 3<= i < 5:
    print(" Bueno almenos lo intente :) ")
    break
  elif  2 <= i <3:
    print(" Bueno espero tener mas suerte la proxima")
    break
  elif  1 <= i < 2:
    print(" Almenos no me odies")
    break
  elif  i < 1:
    print(" vete a la mierda ")
    break
  else:
    print("bueno algo es algo")
    break
```
