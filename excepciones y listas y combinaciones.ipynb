```
# Usando listas como pilas
# Los métodos de lista hacen que resulte muy fácil usar una lista como una pila, donde el último elemento añadido es el primer
# elemento retirado ("último en entrar, primero en salir").
# Para agregar un ítem a la cima de la pila, use append(). Para retirar un ítem de la cima de la pila, use pop()

stack = [3, 4, 5]
stack.append(6)
stack.append(7)
stack # no es necesesario siempre colocar print()
```

```
stack.pop() # Para retirar un ítem de la cima de la pila, use pop()
stack       # cada ves que se compile quita un elemeno mas de la lista existente
```

```
# También es posible usar una lista como una cola, donde el primer elemento añadido es el primer elemento retirado
# Para implementar una cola, usá collections.deque el cual fue diseñado para agregar y sacar de ambas puntas de forma
# rápida.

from collections import deque

queue = deque(["Eric", "John", "Michael"])
queue.append("Terry") # llega Terry
queue
```

```
queue.append("Graham") # llega Graham
queue
```

```
queue.popleft() #saca al primero de la lista y asi sucesivamente
queue
```

```
queue.popleft() #saca al primero de la lista y asi sucesivamente
queue
```

```
# un quivalente
# una convinacion sin repetir valore osea (x1,y1) distint0 (x2,y2)
combs = []
for x in [1,2,3]:
  for y in [3,1,4]:
    if x != y:
       combs.append((x, y))
combs
```
