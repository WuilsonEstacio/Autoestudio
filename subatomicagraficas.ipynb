```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Par√°metros
Vor = 200.0 # En MeV
Vot = 178.0 
Vos = 91.85

e = 1.44 # En Mevfm
kr = 1.487 # en fm-2
kt = 0.639 
ks = 0.465

# Dominio
r = np.linspace(0.2,5, 100)

# Potenciales gaussianos
VR = Vor*np.exp(-kr*r**2) 
VT = -Vot*np.exp(-kt*r**2) 
VS = -Vos*np.exp(-ks*r**2)
```

```
#Interaccion n-n, p-p, n-p
elec = np.array([0.0,1.0,0.0])
Psigma = np.array([-1.0, -1.0, -1.0])
Pr  = np.array([-1.0,-1.0, -1.0] )
taui = np.array([-0.5,0.5,-0.5])
tauj = np.array([-0.5,0.5,0.5])
inte = ['n-n','p-p','n-p']

fig = plt.figure()
ax = fig.add_subplot(111)
ax.grid()
ax.set_ylabel('Potencial V (MeV)')
ax.set_xlabel('Distancia r (fm)')

for i in range(len(Psigma)):
#i = 1
    V1 = (VR + 0.5*(1.0 + Psigma[i])*VT + 0.5*(1.0-Psigma[i])*VS)*(0.5+0.5*Pr[i]) + elec[i]*(0.5*0.5*(1.0+taui[i])*(1.0+tauj[i])*e**2)/(r)
    #V2 = elec[i]*(0.5*0.5*(1.0+taui[i])*(1.0+tauj[i])*e**2)/(r)
    ax.plot(r,V1)
    #ax.plot(r,V2)
    a = inte[i]
    print(a)
    ax.legend([inte[0],inte[1], inte[2]])
fig.savefig("S1.pdf")
```
