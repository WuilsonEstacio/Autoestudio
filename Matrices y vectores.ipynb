```
# matrices

import numpy as np

matrisA=np.array([[1,2,3,4],[3,4,8,3],[9,4,8,23],[4,6,22,53]])
print('matrisA =', matrisA)

matrisB=np.array([[7,33,45,8],[3,4,8,2],[23,14,5,67],[44,6,2,35]])
print('matrisB =', matrisB)
```

```
# suma de matrices
suma=matrisA+matrisB
print(suma)
```

```
# suma de matrices
suma=matrisA-matrisB
print(suma)
```

```
# multiplicacion de matrices
multipli=np.dot(matrisA,matrisB)
print(multipli)
```

```
A=np.linalg.det(matrisA)
print(A)
```

```
np.trace(matrisA)
```
