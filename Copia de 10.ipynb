```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
A = 238.0
c =3.0e8
mpc2 = 938.3  #MeV
mnc2 = 939.6  
```

```python
av = 15.56    
ass = 17.23   
asym = 93.14  
ac = 0.697    
ap = 12.0     
delta = ap/(A**(1.0/2.0))
```

```python
alpha = mnc2 - av + ass/(A**(1.0/3.0)) + asym/4.0
beta = asym + (mnc2 - mpc2)
gamma = asym/A + ac/(A**(1.0/3.0))
Z=  np.linspace(1, 120, 100)
```

```python
M = alpha*A - beta*Z + gamma*Z*Z + delta/(A**(0.5)) 
M = M*1.783e-30/(1.66054e-27) -A
```

```python
fig = plt.figure()   #Creamos una Figura (canvas vacío) 
ax = fig.add_axes([0.15, 0.15, 0.8, 0.8])
ax.plot(Z, M , label="A = 238 ",color='b')
ax.plot(Z1,D, 'go' )

#ax.set_ylim(39.95, 40.02)
ax.set_ylim(0.04, 0.1)
ax.set_xlim(84, 102)
ax.set_ylabel('Exceso de masa (uma)') # Nótese el uso de "set_" al principio del nombre de los métodos
ax.set_xlabel('Número de protones Z')
#ax.set_title('Parábola de masas para Isobars')
ax.grid()
ax.legend()
fig.show()
fig.savefig("10impar.pdf")
```
